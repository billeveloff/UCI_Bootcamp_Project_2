```python
# Import CSV file and create df
file = "test_data.csv"
shotlist_df = pd.read_csv(file)
shotlist_df.head()
```

```python
# Want to drop all rows with no shot data (ie. where Loc_X and Loc_Y are '0')

# createa new dataframe object with the Loc_X and Loc_Y as '0'
removeRows = shotlist_df[(shotlist_df['Loc_X'] == 0) & (shotlist_df['Loc_Y'] == 0)].index
removeRows
```

```python
# Drop this object from our Shotlist_df
shotlist_df.drop(removeRows, inplace=True)
```
