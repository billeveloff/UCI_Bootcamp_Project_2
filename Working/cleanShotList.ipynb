```python
# Import dependices
import pandas as pd
import csv
```

```python
# Import CSV file and create df
file = "shot_list.csv"
shotlist_df = pd.read_csv(file)
shotlist_df.head()
```

```python
## Look at the data types
shotlist_df.dtypes
```

```python
shotlist_df.head()
```

```python
## Create logic that will use '&' on 'Game_Date' and arithmetic operators to create 'Season' column
## with '2003-04' format. This will also create 'NaN' values for all 'Game_Date' outside our range
## We can drop our 'NaN' values using .dropna()
## Finally we can reset the index by using .reset_index(drop=True)
```

```python
## create identical df to test on
shotlist2_df = shotlist_df
shotlist2_df.head()
```

```python
## Using 'Game_Date' column and '&' logic  to create a new 'Season' column
## Only using logic for desired Seasons
## Seasons outside our range will return 'NaN'

shotlist2_df.loc[(shotlist2_df.Game_Date >= 20030418) & (shotlist2_df.Game_Date < 20040430), 'Season'] = '2003-04'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20040418) & (shotlist2_df.Game_Date < 20050430), 'Season'] = '2004-05'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20050418) & (shotlist2_df.Game_Date < 20060430), 'Season'] = '2005-06'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20060418) & (shotlist2_df.Game_Date < 20070430), 'Season'] = '2006-07'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20070418) & (shotlist2_df.Game_Date < 20080430), 'Season'] = '2007-08'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20080418) & (shotlist2_df.Game_Date < 20090430), 'Season'] = '2008-09'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20090418) & (shotlist2_df.Game_Date < 20100430), 'Season'] = '2009-10'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20100418) & (shotlist2_df.Game_Date < 20110430), 'Season'] = '2010-11'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20110418) & (shotlist2_df.Game_Date < 20120430), 'Season'] = '2011-12'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20120418) & (shotlist2_df.Game_Date < 20130430), 'Season'] = '2012-13'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20130418) & (shotlist2_df.Game_Date < 20140430), 'Season'] = '2013-14'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20140418) & (shotlist2_df.Game_Date < 20150430), 'Season'] = '2014-15'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20150418) & (shotlist2_df.Game_Date < 20160430), 'Season'] = '2015-16'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20160418) & (shotlist2_df.Game_Date < 20170430), 'Season'] = '2016-17'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20170418) & (shotlist2_df.Game_Date < 20180430), 'Season'] = '2017-18'
shotlist2_df.loc[(shotlist2_df.Game_Date >= 20180418) & (shotlist2_df.Game_Date < 20190430), 'Season'] = '2018-19'
```

```python
shotlist2_df.head()
```

```python
## We will drop all rows not in our desired range (ie. drop rows with 'NaN' values)
shotlist3_df = shotlist2_df.dropna()
```

```python
## We are going to reset the index and use the 'drop' parameter to ensure the original index is not in a column
shotlist3_df.reset_index(drop=True)
```

```python
## Export our clean shot_list
shotlist3_df.to_csv("master_shot.csv", encoding='utf-8')
```
